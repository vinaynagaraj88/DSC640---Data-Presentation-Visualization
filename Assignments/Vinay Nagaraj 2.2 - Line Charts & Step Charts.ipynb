```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
world_pop = pd.read_excel('world-population.xlsm')
world_pop.head()
```

```python
# Line Chart
plt.plot(world_pop.Year, world_pop.Population)
plt.title('Increase in Population over Years')
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.show()
```

```python
# Step Chart

plt.step(world_pop.Year, world_pop.Population)
plt.title('Increase in Population over Years')
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.show()
```
